corner detection
 The purpose of detecting corners is to track things like motion, do 3D modeling, and recognize objects, shapes, and characters.

```python
import cv2
import numpy as np
img13 = cv2.imread('Pictures/opencv-corner-detection-sample.jpg')
gray = cv2.cvtColor(img13,cv2.COLOR_BGR2GRAY)   #image converted t gray_scale
gray = np.float32(gray)     # gray scale to float32

corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 10)   #GOodFeatureToTrack(image,max_corner_to_detect,quality,minimum_dist_b/w_corners)
corners = np.int0(corners)
for corner in corners:
    x,y = corner.ravel()
    cv2.circle(img13,(x,y),3,255,-1)
    
cv2.imshow('Corner',img13)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

Motion Detection

```python
import numpy as np
import cv2

cap = cv2.VideoCapture('Downloads/people-walking.mp4')
fgbg = cv2.createBackgroundSubtractorMOG2()

while(1):
    ret, frame = cap.read()

    fgmask = fgbg.apply(frame)
 
    cv2.imshow('fgmask',frame)
    cv2.imshow('frame',fgmask)

    if cv2.waitKey(1) & 0xFF == ord('q'):      #waiting for q key 
        break
cap.release()

cv2.destroyAllWindows()
```
